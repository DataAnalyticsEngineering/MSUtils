```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open the HDF5 file
file_path = '../../data/FNOCG_2D.h5'
with h5py.File(file_path, 'r') as h5_file:
    # Load the data from the specified path
    data = h5_file['/feature_extractors/pcf_basis'][:]

# Print the loaded data
print(data.shape)
# Compute min, max, and mean of each mode
min_values = np.min(data, axis=0)
max_values = np.max(data, axis=0)
mean_values = np.mean(data, axis=0)

# Plot the min, max, and mean values for each mode
modes = np.arange(data.shape[1])

plt.figure(figsize=(14, 7))
plt.plot(modes, min_values, label='Min Values')
plt.plot(modes, max_values, label='Max Values')
plt.plot(modes, mean_values, label='Mean Values')
plt.xlabel('Modes')
plt.ylabel('Values')
plt.title('Min, Max, and Mean Values of Each Mode')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(4, 5, figsize=(20, 16))

vmin = np.min(data[:, :10])
vmax = np.max(data[:, :10])

for i in range(20):
    mode_reshaped = data[:, i].reshape(400, 400)
    ax = axes[i // 5, i % 5]
    im = ax.imshow(mode_reshaped, cmap='viridis', vmin=vmin, vmax=vmax)
    ax.set_title(f'Mode {i}')
    ax.axis('off')

cbar = fig.colorbar(im, ax=axes, orientation='horizontal', fraction=0.05, pad=0.05)
plt.tight_layout()
plt.show()
```
